 Load all the required libraries.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

 Load dataset:

```python
dataset = pd.read_csv("D:/ML/archive/heart.csv")
```

To see the patients who have heart disease or not.

To distinguish the target variable between male and female.

```python
sns.countplot(x='sex', data=dataset, palette="bwr")
plt.xlabel("Sex (0 = female, 1= male)")
plt.show()
```

Relationship between Maximum Heart Rate and Age and visualizing it with scatter plot.

```python
plt.scatter(x=dataset.age[dataset.target==1], y=dataset.thalach[(dataset.target==1)], c='red')
plt.scatter(x=dataset.age[dataset.target==0], y=dataset.thalach[(dataset.target==0)], c = 'black')
plt.legend(["Disease", "Not Disease"])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.show()
```

Labeling the dataset

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,13].values
```

Splitting 75% of the data to the training while 25% of the data to test the accuracy of the prediction.

```python
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.25, random_state= 0)
```

Normalize the data 

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

Making a K-NN model and train it with the training data and K = 5 also finding out its accuracy.

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier = classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
#check accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```
